### SARSA (on-policy TD control) for estimating $Q \approx q_*$

- Initialize $Q(s, a)$, for all $s \in \mathcal{S}, a \in \mathcal{A}(s)$, arbitrarily, and $Q($ terminal-state,$\cdot)=0$ 

- Repeat (for each episode):
    - Initialize $S$ 
    - Choose $A$ from $S$ using policy derived from $Q$ (e.g., $\epsilon$ -greedy $)$
    - Repeat (for each step of episode):
        - Take action $A$, observe $R, S^{\prime}$ 
        - Choose $A^{\prime}$ from $S^{\prime}$ using policy derived from $Q$ (e.g., $\epsilon$ -greedy)
        - $Q(S, A) \leftarrow Q(S, A)+\alpha\left[R+\gamma Q\left(S^{\prime}, A^{\prime}\right)-Q(S, A)\right]$
        - $S \leftarrow S^{\prime}$
        - $A \leftarrow A^{\prime}$
    - until $S$ is terminal

#### SARSA (Cliff Walking Environment / Agent) 구현

> `CliffWalking_SARSA` class에 `environment`, `agent` 같이 구현

- $\alpha = 0.5$ (alpha)
- $\epsilon = 0.1$ (epsilon)
- $\gamma = 1$ (discount_factor)

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(seed=21)
```

```python
class CliffWalking_SARSA():
    def __init__(self, state, alpha = 0.5, discount_factor = 1, epsilon = 0.1):
        self.state = state                  # np.zeros([7, 10]) 
        self.q_table = np.zeros([4, 4, 12]) # 갯수 : action * state
        self.terminated = False             # Terminal 도착시 다음 Episode로 돌도록 조건 설정
        self.alpha = alpha                  # step size
        self.actions = np.arange(4)         # (0 : up, 1 : down, 2 : reft, 3 : right)
        self.discount_factor = discount_factor
        self.epsilon = epsilon              
        self.row = 3                        # Initial location (state) is row (y)
        self.column = 0                     # Initial location (state) is column (x)
    
    
    # epsilon greedy policy에 따라 action 반환
    def get_action_with_e_greedy(self, state):
        p = np.random.random()              # 0 ~ 1의 범위를 가지는 uniform distribution에서 p 추출
        if p < self.epsilon:
            # 무작위 행동 반환
            action = np.random.choice(len(self.actions))
            return action
        else:
            
            # 방법 1 : 가장 큰 값이 여러개인 경우 첫번째 indew만 출력
            state_action = self.q_table[:, state[0], state[1]]
            action = np.argmax(state_action)
            
            # 방법 2 : 가장 큰 값이 여러개인 경우 random하게 뽑음
            # 큐함수에 따른 행동 반환
#             state_action = self.q_table[:, state[0], state[1]]
#             action = np.random.choice(np.where(state_action == state_action.max())[0])
            return action
    
    def take_action(self, action):
        '''
        args[0] = action
        return reward
        '''
        # up
        if action == 0:
            self.row -= 1

        # down
        elif action == 1:
            self.row += 1

        # left
        elif action == 2:
            self.column -= 1
        
        # right
        elif action == 3:
            self.column += 1
            
        
        # 이동 후 왼쪽 벽에 부딪친 경우 state, reward
        if self.column in [-1]:
            self.column = 0
            reward = -1
        # 이동 후 오른쪽 벽에 부딪친 경우 state, reward
        elif self.column in [12]:
            self.column = 11
            reward = -1
        # 이외의 경우 state, reward
        else:
            reward = -1
            
        # 이동 후 위 벽에 부딪친 경우 state, reward
        if self.row in [-1]:
            self.row = 0
            reward = -1
        # start point에서 아래로 이동하는 경우 state, reward 
        elif (self.row, self.column) == (4, 0):
            # 다시 start point 설정!
            self.row, self.column = 3, 0
            reward = -1
        # 초기 start point에 도달하는 경우 state, reward
        elif (self.row, self.column) == (3, 0):
            # 이전 상태로 되돌림
            self.row, self.column = 2, 0
            reward = -1
        # The cliff 벽에 부딪친 경우 Terminal 종료 (자동으로 state 초기화) 및 reward = -100
        elif (self.row, self.column) in zip([3]*10, list(np.arange(1,11))):
            self.terminated = True
            reward = -100
            # print('Case that fell on The Cliff')
        # Terminal에 도달한 경우 self.terminated를 True로 설정            
        elif self.state[self.row][self.column] == 2.0:
            self.terminated = True
            # print('Terminated')
            reward = -1
            
        return reward
    
    
    # <s, a, r, s', a'>의 샘플로부터 큐 함수 업데이트
    def learn_with_SARSA(self, state, action, reward, next_state, next_action):
        # 현재 state, action의 q value
        current_state_q = self.q_table[action, state[0], state[1]]
        # 다음 state, action의 q value        
        next_state_q = self.q_table[next_action, next_state[0], next_state[1]]
        # update q value
        new_q = (current_state_q + self.alpha * (reward + self.discount_factor * next_state_q - current_state_q))
        self.q_table[action, state[0], state[1]] = new_q
```

#### SARSA (Cliff Walking) 실험

- episode = 500
- reward = -1 or -100

```python
# state 정의
states = np.zeros([4, 12]) # height x width 크기의 state
states[3][0] = 1           # start value = 1
states[3][11] = 2          # terminal check value = 2

# WindyGridWorld 초기화
CW_SARSA = CliffWalking_SARSA(state = states)

episodes = 500
step = 1


# time steps 및 episodes 이력 저장

sarsa_episode_list = []
sarsa_sum_of_during_episode_rewards_list = []

# 설정된 episode 만큼 iteration
for episode in range(episodes):
    
    # Episode 마다 WindyGridWorld 상태를 초기화 (반드시 q_table은 유지)
    CW_SARSA.row, CW_SARSA.column = 3, 0
    CW_SARSA.terminated = False
    # Episode에 따른 reward 누적 log
    temp_reward_sum = 0
    
    # step에 따른 큐함수 update 과정 (Terminal에 도착하면 windy_grid_world.terminated = True로 while 탈출)
    while not CW_SARSA.terminated:
        
        # current state
        state = [CW_SARSA.row, CW_SARSA.column]
        # current action
        action = CW_SARSA.get_action_with_e_greedy(state)
        # current state에서 action으로 인한 state 이동 후 reward 출력
        reward = CW_SARSA.take_action(action)
        # next_state
        next_state = [CW_SARSA.row, CW_SARSA.column]
        # next_action
        next_action = CW_SARSA.get_action_with_e_greedy(next_state)
        # wgr.take_action(next_action)
        
        # <s, a, r, s', a'>의 샘플로부터 SARSA algorithms에 의한 큐함수를 업데이트
        CW_SARSA.learn_with_SARSA(state, action, reward, next_state, next_action)
        print(next_state)
#         print("step : {} \nstate : ({},{}), action : {}".format(step, state[0], state[1], action))
#         print("next_state : ({},{}), next_action : {}".format(next_state[0], next_state[1], next_action))
#         print('-----------------------------------')
        
        step += 1
        temp_reward_sum += reward
        
    
    # episode 및 reward sum log(이력) 저장
    sarsa_episode_list.append(episode)
    sarsa_sum_of_during_episode_rewards_list.append(temp_reward_sum)
```

### Q-Learning (off-policy) for estimating $\pi \approx \pi_*$

- Initialize $Q(s, a)$, for all $s \in \mathcal{S}, a \in \mathcal{A}(s)$, arbitrarily, and $Q($ terminal-state,$\cdot)=0$ 

- Repeat (for each episode):
    - Initialize $S$
    - Repeat (for each step of episode):
        - Choose $A$ from $S$ using policy derived from $Q$ (e.g., $\epsilon$ -greedy $)$    
        - Take action $A$, observe $R, S^{\prime}$ 
        - $Q(S, A) \leftarrow Q(S, A)+\alpha\left[R+\gamma \text{ argmax}_{a} Q\left(S^{\prime}, a\right)-Q(S, A)\right]$
        - $S \leftarrow S^{\prime}$
    - until $S$ is terminal

#### Q-Learning (Cliff Walking Environment / Agent)  구현

```python
class CliffWalking_Qlearning():
    def __init__(self, state, alpha = 0.5, discount_factor = 1, epsilon = 0.1):
        self.state = state                  # np.zeros([7, 10]) 
        self.q_table = np.zeros([4, 4, 12]) # 갯수 : action * state
        self.terminated = False             # Terminal 도착시 다음 Episode로 돌도록 조건 설정
        self.alpha = alpha                  # step size
        self.actions = np.arange(4)         # (0 : up, 1 : down, 2 : reft, 3 : right)
        self.discount_factor = discount_factor
        self.epsilon = epsilon              
        self.row = 3                        # Initial location (state) is row (y)
        self.column = 0                     # Initial location (state) is column (x)
    
    
    # epsilon greedy policy에 따라 action 반환
    def get_action_with_e_greedy(self, state):
        p = np.random.random()              # 0 ~ 1의 범위를 가지는 uniform distribution에서 p 추출
        if p < self.epsilon:
            # 무작위 행동 반환
            action = np.random.choice(len(self.actions))
            return action
        else:
            
            # 방법 1 : 가장 큰 값이 여러개인 경우 첫번째 indew만 출력
            state_action = self.q_table[:, state[0], state[1]]
            action = np.argmax(state_action)
            
            # 방법 2 : 가장 큰 값이 여러개인 경우 random하게 뽑음
            # 큐함수에 따른 행동 반환
#             state_action = self.q_table[:, state[0], state[1]]
#             action = np.random.choice(np.where(state_action == state_action.max())[0])
            return action
    
    def take_action(self, action):
        '''
        args[0] = action
        return reward
        '''
        # up
        if action == 0:
            self.row -= 1

        # down
        elif action == 1:
            self.row += 1

        # left
        elif action == 2:
            self.column -= 1
        
        # right
        elif action == 3:
            self.column += 1
            
        
        # 이동 후 왼쪽 벽에 부딪친 경우 state, reward
        if self.column in [-1]:
            self.column = 0
            reward = -1
        # 이동 후 오른쪽 벽에 부딪친 경우 state, reward
        elif self.column in [12]:
            self.column = 11
            reward = -1
        # 이외의 경우 state, reward
        else:
            reward = -1
            
        # 이동 후 위 벽에 부딪친 경우 state, reward
        if self.row in [-1]:
            self.row = 0
            reward = -1
        # start point에서 아래로 이동하는 경우 state, reward 
        elif (self.row, self.column) == (4, 0):
            # 다시 start point 설정!
            self.row, self.column = 3, 0
            reward = -1
        # 초기 start point에 도달하는 경우 state, reward
        elif (self.row, self.column) == (3, 0):
            # 이전 상태로 되돌림
            self.row, self.column = 2, 0
            reward = -1
        # The cliff 벽에 부딪친 경우 Terminal 종료 (자동으로 state 초기화) 및 reward = -100
        elif (self.row, self.column) in zip([3]*10, list(np.arange(1,11))):
            self.terminated = True
            reward = -100
            # print('Case that fell on The Cliff')
        # Terminal에 도달한 경우 self.terminated를 True로 설정            
        elif self.state[self.row][self.column] == 2.0:
            self.terminated = True
            # print('Terminated')
            reward = -1
            
        return reward
    
    
    # <s, a, r, s'>의 샘플로부터 큐 함수 업데이트
    def learn_with_Qlearning(self, state, action, reward, next_state):
        # 현재 state, action의 q value
        current_state_q = self.q_table[action, state[0], state[1]]
        # 벨만 최적 방정식을 사용한 큐 함수 업데이트
        new_q = (current_state_q + self.alpha * (reward + self.discount_factor * max(self.q_table[:, next_state[0], next_state[1]]) - current_state_q))
        self.q_table[action, state[0], state[1]] = new_q
```

#### Q-Learning (Cliff Walking) 실험

```python
# state 정의
states = np.zeros([4, 12]) # height x width 크기의 state
states[3][0] = 1           # start value = 1
states[3][11] = 2          # terminal check value = 2

# WindyGridWorld 초기화
CW_Q = CliffWalking_Qlearning(state = states)

episodes = 500
step = 1


# time steps 및 episodes 이력 저장

q_episode_list = []
q_sum_of_during_episode_rewards_list = []

# 설정된 episode 만큼 iteration
for episode in range(episodes):
    
    # Episode 마다 WindyGridWorld 상태를 초기화 (반드시 q_table은 유지)
    CW_Q.row, CW_Q.column = 3, 0
    CW_Q.terminated = False
    # Episode에 따른 reward 누적 log
    temp_reward_sum = 0
    
    # step에 따른 큐함수 update 과정 (Terminal에 도착하면 windy_grid_world.terminated = True로 while 탈출)
    while not CW_Q.terminated:
        
        # current state
        state = [CW_Q.row, CW_Q.column]
        # current action
        action = CW_Q.get_action_with_e_greedy(state)
        # current state에서 action으로 인한 state 이동 후 reward 출력
        reward = CW_Q.take_action(action)
        # print(reward)
        # next_state
        next_state = [CW_Q.row, CW_Q.column]
        
        # <s, a, r, s'>의 샘플로부터 SARSA algorithms에 의한 큐함수를 업데이트
        CW_Q.learn_with_Qlearning(state, action, reward, next_state)
#         print("step : {} \nstate : ({},{}), action : {}".format(step, state[0], state[1], action))
#         print('-----------------------------------')
        
        step += 1
        temp_reward_sum += reward
        
    
    # episode 및 reward sum log(이력) 저장
    q_episode_list.append(episode)
    q_sum_of_during_episode_rewards_list.append(temp_reward_sum)
```

> **SARSA** : SARSA에서는 충분한 탐험을 하기 위해 $\epsilon$-greedy policy를 사용하여 현재 상태의 큐함수를 다음과 같이 업데이트 하는것을 생각해보자. $Q(S, A) \leftarrow Q(S, A)+\alpha\left[R+\gamma Q\left(S^{\prime}, A^{\prime}\right)-Q(S, A)\right]$ 이때, $Q(S',A')$의 결과로부터 현재의 큐함수를 업데이트하게 된다. 만일, reward가 작은 state를 만난다면, q 값을 낮추게 만드는데, 이는 탐험에 대한 정책을 막을 수 있는 단점(문제점) 발생하게 된다. 즉 탐험을 하지 못하기 대문에 최적 정책을 학습하지 못할 가능성이 발생한다. (이 예제에서는 SARSA에 대한 단점이 reward 관점에서는 바로 보이지는 않으며, episode가 증가하면서 reward의 합이 커지다가 `-20` 근처에서 수렴하는것을 확인할 수 있다. 
> 
> **Q-learning** : Q-learning은 SARSA의 단점을 해결하기 위해, 행동하는 정책과 학습하는 정책을 따로 분리한 알고리즘이다. 즉, 실제 환경에서 행동을 하는 정책과 큐함수를 업데이트할 때 사용하는 정책이 다르기 때문에 SARSA에서의 탐험을 못하는 문제를 해결할 수 있다. 위 예제로 그러한 문제를 결과로 Q-learning이 rewards가 더 낮아지는 것을 바로 비교하기에는 어렵지만, Q-learning의 특성 상, episode가 증가하면서 `SARSA`보다는 탐험을 보다 많이 하기 때문에(e.g. episode 마다 cliff에 닿음), reward가 더 낮은 빈도수가 많은것을 확인할 수 있다.

> 결론 : SARSA 보다 Q-learning이 탐험을 더 많이 하므로, Episode에 따른 reward가 낮은 경향을 보이고 있다!
