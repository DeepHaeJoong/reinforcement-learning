### Episodic Semi-gradient SARSA for estimating $\hat{q} \approx q_*$

- Input : a differentiable action-value function parameterization $\hat{q} : S \times A \times \mathbb{R}^d \rightarrow \mathbb{R}$
- Algorithm parameters : step size $\alpha \gt 0$, small $\epsilon \gt 0$
- Initialize value-function weights $\mathbf{w} \in \mathbb{R}^d$ arbitrarily (e.g., $\mathbf{w}=0$)

- Loop for each episode:
    - $S, A \leftarrow$ initial state and action of episode (e.g., $\epsilon$ -greedy $)$
    - Loop for each step of episode:
        - Take action $A$, observer $R, S'$
        - If $S'$ is terminal:
            - $\mathbf{w} \leftarrow \mathbf{w} + \alpha \left[R-\hat{q}(S,A,\mathbf{w})\right] \nabla \hat{q}(S,A,\mathbf{w})$
            - Go to next episode
        - Choose $A'$ as a function of $\hat{q}(S', \cdot, \mathbf{w})$ (e.g. $\epsilon$ -greedy)
        - $\mathbf{w} \leftarrow \mathbf{w} + \alpha \left[R + \gamma \max_a \hat{q}(S',a,\mathbf{w}) -\hat{q}(S,A,\mathbf{w})\right] \nabla \hat{q}(S,A,\mathbf{w})$
        - $S \leftarrow S'$
        - $A \leftarrow A'$

### Semi-gradient SARSA (Mountain Car Task Environment / Agent) 구현


<img src="https://www.researchgate.net/profile/Matthew-Luciw/publication/224261111/figure/fig1/AS:302696662880256@1449179916353/Mountain-Car-task-In-the-standard-version-the-cars-position-and-velocity-are-provided.png" alt="drawing" width="300"/>



#### state, action, reward

- STATES (SITUATIONS) : car's `position` and `velocity`
    - position : $x_t$
        - $x_{t+1} = \text{bound}[x_{t} + \dot{x}_{t+1}]$
    - velocity : $\dot{x}_{t}$
        - $\dot{x}_{t+1} = \text{bound}[\dot{x}_{t} + 0.001A_t -0.0025\text{cos}(3x_t)]$
        
- ACTIONS : $A_t$ 
    - forward(`+1`)
    - reverse(`-1`)
    - none(`0`)
    
- REWARDS : always `–1` until car reaches the goal

#### bound

- bound of position : $ -1.2 \leq x_{t+1} \leq 0.5$
- bound of velocity : $ -0.07 \leq \dot{x}_{t+1} \leq 0.07$

#### Initialize position, velocity

- random position $x_t \in [-0.6, -0.4]$ by uniform distribution
- velocity : 0


#### hyperparameter

- $\alpha = 0.5$ (alpha)
- $\epsilon = 0.1$ (epsilon)
- $\gamma = 1$ (discount_factor)

#### grid-tiling

- used 8 tilings, each tile covering 1/8 th of the bounded distance in each dimension.
- $\hat{q}(s,a,\mathbf{w}) = \mathbf{w}^T\mathbf{x}(s,a)= \sum_{i=1}^{d}w^i \cdot x_i(s,a)$
- `iht=IHT(4096)` and `tiles(iht,8,[8*x/(0.5+1.2),8*xdot/(0.07+0.07)],[A])`
    - 교재 page 246 하단에 위 기법을 사용하라는 언급이 있어 그대로 [참고](http://incompleteideas.net/tiles/tiles3.py-remove)하였으며, 해당 code가 class로 구현되어 있어 사용하였습니다.

```python
import numpy as np
from collections import namedtuple

# 3D plot을 위한 library
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

import sys
import time
import timeit
```

```python
'''
Tile Coding Software version 3.0beta
by Rich Sutton
based on a program created by Steph Schaeffer and others
External documentation and recommendations on the use of this code is available in the 
reinforcement learning textbook by Sutton and Barto, and on the web.
These need to be understood before this code is.

This software is for Python 3 or more.

This is an implementation of grid-style tile codings, based originally on
the UNH CMAC code (see http://www.ece.unh.edu/robots/cmac.htm), but by now highly changed. 
Here we provide a function, "tiles", that maps floating and integer
variables to a list of tiles, and a second function "tiles-wrap" that does the same while
wrapping some floats to provided widths (the lower wrap value is always 0).

The float variables will be gridded at unit intervals, so generalization
will be by approximately 1 in each direction, and any scaling will have 
to be done externally before calling tiles.

Num-tilings should be a power of 2, e.g., 16. To make the offsetting work properly, it should
also be greater than or equal to four times the number of floats.

The first argument is either an index hash table of a given size (created by (make-iht size)), 
an integer "size" (range of the indices from 0), or nil (for testing, indicating that the tile 
coordinates are to be returned without being converted to indices).
'''

basehash = hash

class IHT:
    "Structure to handle collisions"
    def __init__(self, sizeval):
        self.size = sizeval                        
        self.overfullCount = 0
        self.dictionary = {}

    def __str__(self):
        "Prepares a string for printing whenever this object is printed"
        return "Collision table:" + \
               " size:" + str(self.size) + \
               " overfullCount:" + str(self.overfullCount) + \
               " dictionary:" + str(len(self.dictionary)) + " items"

    def count (self):
        return len(self.dictionary)
    
    def fullp (self):
        return len(self.dictionary) >= self.size
    
    def getindex (self, obj, readonly=False):
        d = self.dictionary
        if obj in d: return d[obj]
        elif readonly: return None
        size = self.size
        count = self.count()
        if count >= size:
            if self.overfullCount==0: print('IHT full, starting to allow collisions')
            self.overfullCount += 1
            return basehash(obj) % self.size
        else:
            d[obj] = count
            return count

def hashcoords(coordinates, m, readonly=False):
    if type(m)==IHT: return m.getindex(tuple(coordinates), readonly)
    if type(m)==int: return basehash(tuple(coordinates)) % m
    if m==None: return coordinates

from math import floor, log
from itertools import zip_longest

def tiles (ihtORsize, numtilings, floats, ints=[], readonly=False):
    """returns num-tilings tile indices corresponding to the floats and ints"""
    qfloats = [floor(f*numtilings) for f in floats]
    Tiles = []
    for tiling in range(numtilings):
        tilingX2 = tiling*2
        coords = [tiling]
        b = tiling
        for q in qfloats:
            coords.append( (q + b) // numtilings )
            b += tilingX2
        coords.extend(ints)
        Tiles.append(hashcoords(coords, ihtORsize, readonly))
    return Tiles

def tileswrap (ihtORsize, numtilings, floats, wrapwidths, ints=[], readonly=False):
    """returns num-tilings tile indices corresponding to the floats and ints, wrapping some floats"""
    qfloats = [floor(f*numtilings) for f in floats]
    Tiles = []
    for tiling in range(numtilings):
        tilingX2 = tiling*2
        coords = [tiling]
        b = tiling
        for q, width in zip_longest(qfloats, wrapwidths):
            c = (q + b%numtilings) // numtilings
            coords.append(c%width if width else c)
            b += tilingX2
        coords.extend(ints)
        Tiles.append(hashcoords(coords, ihtORsize, readonly))
    return Tiles
```

### Environment : MountainCarEnv 구현

```python
class MountainCarEnv():
    """
    Environment of Mountain Car
    args : action
    output : 
    """
    def __init__(self):
        self.terminated = False
        self.position = np.random.uniform(-0.6, -0.4) # position : 초기값 (random position [-0.6, -0.4] by uniform distribution)
        self.velocity = 0
        self.position_lower_bound = -1.2
        self.position_high_bound = 0.5
        self.velocity_lower_bound = -0.07
        self.velocity_high_bound = 0.07              
        
        
    def reset(self):
        """
        environment(position, velocity) reset
        args : None
        output : [position, velocity]
        """
        self.position = np.random.uniform(-0.6, -0.4)  # position : 초기값 (random position [-0.6, -0.4] by uniform distribution)
        self.velocity = 0
        self.terminated = False
        return [self.position, self.velocity]
        
    def step(self, action):
        """
        description : 현재 state에서 action에 따른 다음 state (velocity, position update)
        args : action [-1, 0, 1]
        return : next_state, reward, terminated 
        """
        reward = -1

        # velocity update
        new_velocity = self.velocity + 0.001*action - 0.0025*np.cos(3*self.position)
        new_velocity = min(max(self.velocity_lower_bound, new_velocity), self.velocity_high_bound) # new_velocity에 대한 bound 적용
        self.velocity = new_velocity
        
        # position update
        new_position = self.position + new_velocity
        self.position = min(max(self.position_lower_bound, new_position), self.position_high_bound)
        
        # position이 -1.2 와 같으면 velocity = 0 으로 변경
        if self.position == self.position_lower_bound:
            self.velocity = 0.0
        
        # position 0.5를 만족하면 terminated = True
        if self.position == self.position_high_bound:
            self.terminated = True
        
        next_state = [self.position, self.velocity]
        terminated = self.terminated
        
        return next_state, reward, terminated
```

```python
# MountainCarEnv 작동 확인
env = MountainCarEnv()
env.reset()
```

### Agent : Q (s,a)에 대한 linear approximator 구현

```python
class QApproximator():
    """
    Linear action-value (q-value) function approximator for 
    semi-gradient methods with state-action featurization via tile coding. 
    """
    def __init__(self, num_tilings=8, max_size=4096, step_size=0.5, discount_factor=1, trace=False):
        self.num_tilings = num_tilings
        self.max_size = max_size
        
        # Step size is interpreted as the fraction of the way we want 
        # to move towards the target. To compute the learning rate alpha,
        # scale by number of tilings.
        self.alpha = step_size / num_tilings
        
        # Initialize index hash table (IHT) for tile coding.
        # This assigns a unique index to each tile up to max_size tiles.
        # Ensure max_size >= total number of tiles (num_tilings x tiling_dim x tiling_dim)
        # to ensure no duplicates.
        self.iht = IHT(max_size)
        
        # Initialize weights
        self.weights = np.zeros(max_size)
        
        # Tilecoding software partitions at integer boundaries, so must rescale
        # position and velocity space to span tiling_dim x tiling_dim region.
        self.position_scale = num_tilings / (0.5-(-1.2))
        self.velocity_scale = num_tilings / (0.7-(-0.7))
        
        self.discount_factor = discount_factor
        self.trace=trace
    
    # tile coding
    def featurize_state_action(self, state, action):
        """
        Returns the featurized representation for a 
        state-action pair.
        """
        featurized = tiles(self.iht, self.num_tilings, 
                           [self.position_scale * state[0], 
                            self.velocity_scale * state[1]], 
                           [action])

        return featurized
    
    
    def predict(self, s, a=None):
        """
        Predicts q-value(s) using linear FA.
        If action a is given then returns prediction
        for single state-action pair (s, a).
        """
        if a is None:
            features = [self.featurize_state_action(s, i) for i in range(-1,2,1)]
        else:
            features = [self.featurize_state_action(s, a)]
        
        return [np.sum(self.weights[f]) for f in features]
                
    
    def update(self, s, a, target):
        """
        Updates the estimator parameters
        for a given state and action towards
        the target using the gradient update rule.
        """
        features = self.featurize_state_action(s, a)
        print("s : {}".format(s))
        print("a : {}".format(a))        
        estimation = np.sum(self.weights[features])  # Linear FA
        target_error = (target - estimation)
        
        print("estimation : {}".format(estimation))
        print("target : {}".format(target))
        print("target_error : {}".format(target_error))
        
        self.weights[features] += self.alpha * target_error
                
```

### $\epsilon-$ greedy policy 구현

```python
# epsilon greedy policy에 따라 action 반환 
def get_action_with_e_greedy(q_values, epsilon=0):
    """
    Creates an epsilon-greedy policy based on a 
    given q-value approximator and epsilon.
    args: q_values, epsilon
    return : action (-1, 0, 1) 
    """
    actions = [-1, 0, 1]
    p = np.random.random()  # 0 ~ 1의 범위를 가지는 uniform distribution에서 p 추출
    if p < epsilon:
        # 무작위 행동 반환 (-1, 0, 1)
        action = np.random.choice(actions)
        return action
    else:
        # 가장 큰 값에 해당하는 action의 index (-1, 0, 1) 출력  
        action = np.argmax(q_values) - 1
        return action
```

```python
def make_epsilon_greedy_policy(estimator, epsilon, num_actions):
    """
    Creates an epsilon-greedy policy based on a 
    given q-value approximator and epsilon.    
    """
    def policy_fn(observation):      
        action_probs = np.ones(num_actions, dtype=float) * epsilon / num_actions
        q_values = estimator.predict(observation)
        best_action_idx = np.argmax(q_values)
        action_probs[best_action_idx] += (1.0 - epsilon)
        return action_probs
    return policy_fn
```

### Semi-gradient SARSA 구현

```python
def n_step_semi_gradient_sarsa(n, env=MountainCarEnv(), estimator=QApproximator(), gamma=1.0, epsilon=0.1):
    """
    n = 1 로 설정하고 진행 
    """
    # env = MountainCarEnv() 선언
    env = env    
    
    # Reset the environment and pick the first action by epsilon-greedy policy
    state = env.reset()
    q_values = estimator.predict(state)
    action = get_action_with_e_greedy(q_values)
    
    # track previous position, velocity, action and reward
    states = [state]
    actions = [action]
    rewards = [0.0]
    
    # Step through episode
    T = float('inf')
    step = 0
    while not env.terminated:
        # go to next time step        
        if step < T:
            # take current action and go to the new state
            next_state, reward, terminated = env.step(action)
            states.append(next_state)
            rewards.append(reward)
            
            if terminated:
                T = step + 1
    
            else:
                # Take next state / action
                next_q_values = estimator.predict(next_state)
                next_action = get_action_with_e_greedy(next_q_values)
                actions.append(next_action)
        
        # Specifies state to be updated
        update_time = step + 1 - n
        if update_time >= 0:
            # Build target
            target = 0
            # calculate corresponding rewards
            for i in range(update_time + 1, min(T, update_time + n) + 1):
                target += np.power(gamma, i - update_time - 1) * rewards[i]
            # add estimated state action value to the return
            if update_time + n < T:
                next_q_values = estimator.predict(states[update_time + n])
                print("states[update_time + n] : {}".format(states[update_time + n]))
                print("next_q_values : {}".format(next_q_values))
                print(actions)
                print("actions[update_time + n] : {}".format(actions[update_time + n]))
                target += next_q_values[actions[update_time + n]]

            estimator.update(states[update_time], actions[update_time], target)
        
        if update_time == T - 1:       
            break
        
        
        step += 1
        state = next_state
        action = next_action
    
    returns = np.sum(rewards)
    
    return step, returns
```

```python
RunStats = namedtuple('RunStats', ['algorithm', 'steps', 'returns'])
```

```python
def run(algorithm, num_episodes=500, **algorithm_kwargs):
    """
    Runs algorithm over multilple episodes and logs
    for each episode the complete return (G_t) and the
    number of steps taken.
    """
    stats = RunStats(
        algorithm=algorithm, 
        steps=np.zeros(num_episodes), 
        returns=np.zeros(num_episodes))
    
    algorithm_fn = globals()[algorithm]
    
    for i in range(num_episodes):
        episode_steps, episode_return = algorithm_fn(**algorithm_kwargs)
        stats.steps[i] = episode_steps
        stats.returns[i] = episode_return
        sys.stdout.flush()
        print("\rEpisode {}/{} Return {}".format(
            i + 1, num_episodes, episode_return), end="")
    return stats
```

```python
def plot_cost_to_go(env, estimator, num_partitions=50):
    """
    Plots -Q(s, a_max) for each state s=(position, velocity) 
    in the environment where a_max is the maximising action 
    from s according to our q-value estimator Q.
    The state-space is continuous hence we first discretise 
    it into num_partitions partitions in each dimension. 
    """
    x = np.linspace(env.position_lower_bound, env.position_high_bound, num=num_partitions)
    y = np.linspace(env.velocity_lower_bound, env.velocity_high_bound, num=num_partitions)
    X, Y = np.meshgrid(x, y)
    Z = np.apply_along_axis(
        lambda obs: -np.max(estimator.predict(obs)), 2, np.stack([X, Y], axis=2))
    
    fig = plt.figure(figsize=(10, 5))
    ax = plt.axes(projection='3d')
    ax.set_facecolor('white')
    #ax.scatter3D(X, Y, Z, cmap='Greens')
    ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='viridis', edgecolor='none')
    ax.set_xlabel('Position', fontsize=15)
    ax.set_ylabel('Velocity', fontsize=15)
    ax.set_title("\"Cost To Go\" Function", fontsize=20)

    plt.show()
```

```python
step_size = 0.5  # Fraction of the way we want to move towards target
n = 1  # Level of bootstrapping (set to intermediate value)
num_episodes = 1

estimator_n = QApproximator(num_tilings=8, max_size=4096, step_size=0.5, discount_factor=1, trace=False)

start_time = timeit.default_timer()
run_stats_n = run('n_step_semi_gradient_sarsa', num_episodes, n=n, env=MountainCarEnv(), estimator=estimator_n)
elapsed_time = timeit.default_timer() - start_time

plot_cost_to_go(MountainCarEnv(), estimator_n)
print('{} episodes completed in {:.2f}s'.format(num_episodes, elapsed_time))
```
